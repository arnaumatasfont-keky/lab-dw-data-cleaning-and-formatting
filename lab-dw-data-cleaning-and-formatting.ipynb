## Exercise 2: Cleaning invalid Values

The dataset contains columns with inconsistent and incorrect values that could affect the accuracy of our analysis. Therefore, we need to clean these columns to ensure that they only contain valid data.

Note that this exercise will focus only on cleaning inconsistent values and will not involve handling null values (NaN or None).

*Hint*:
- *Gender column contains various inconsistent values such as "F", "M", "Femal", "Male", "female", which need to be standardized, for example, to "M" and "F".*
- *State abbreviations be can replaced with its full name, for example "AZ": "Arizona", "Cali": "California", "WA": "Washington"*
- *In education, "Bachelors" could be replaced by "Bachelor"*
- *In Customer Lifetime Value, delete the `%` character*
- *In vehicle class, "Sports Car", "Luxury SUV" and "Luxury Car" could be replaced by "Luxury"*

```python
# Your code here
columnes_check = ['gender', 'state', 'education', 'vehicle_class', 'customer_lifetime_value']
for col in columnes_check:
    print(f"--- Valors únics a la columna: {col} ---")
    print(df[col].value_counts(dropna=False)) # dropna=False per veure també quants nuls hi ha
    print("\n")
```

```python
gender_map = {
    "Male": "M",
    "female": "F",
    "Femal": "F",
    "F": "F",
    "M": "M"
}
df['gender'] = df['gender'].replace(gender_map)

state_map = {
    "Cali": "California",
    "WA": "Washington",
    "AZ": "Arizona"
}
df['state'] = df['state'].replace(state_map)

df['education'] = df['education'].replace("Bachelors", "Bachelor")

luxury_map = {
    "Sports Car": "Luxury",
    "Luxury SUV": "Luxury",
    "Luxury Car": "Luxury"
}
df['vehicle_class'] = df['vehicle_class'].replace(luxury_map)

df['customer_lifetime_value'] = df['customer_lifetime_value'].str.replace('%', '')
df['customer_lifetime_value'] = pd.to_numeric(df['customer_lifetime_value'], errors='coerce')

print("Gender:", df['gender'].unique())
print("State:", df['state'].unique())
print("Vehicle Class:", df['vehicle_class'].unique())
print("Tipus CLV:", df['customer_lifetime_value'].dtype)
```

```python
# Your code here
nuls_per_columna = df.isnull().sum()
print(nuls_per_columna)
```

```python
df = df.dropna(subset=['customer'])
```

```python
numeric_cols = df.select_dtypes(include=['number']).columns
for col in numeric_cols:
    df[col] = df[col].fillna(df[col].mean())
```

```python
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    df[col] = df[col].fillna('Unknown')
```

```python
for col in numeric_cols:
    df[col] = df[col].astype(int)
```

```python
print("Nuls after cleaning:")
print(df.isnull().sum().sum()) # Hauria de donar 0
print("\nData types after change:")
print(df.dtypes)
```

```python
high_claim_threshold = df_cleaned['total_claim_amount'].quantile(0.75)
low_clv_threshold = df_cleaned['customer_lifetime_value'].quantile(0.25)
```

```python
high_risk_customers = df_cleaned[
    (df_cleaned['total_claim_amount'] > high_claim_threshold) & 
    (df_cleaned['customer_lifetime_value'] < low_clv_threshold)]
```

```python
analysis_summary = high_risk_customers[['total_claim_amount', 'customer_lifetime_value']].describe()
display(analysis_summary)
```
